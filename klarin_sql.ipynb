## Task 2
- Find the number of user reviews and the average rating for each book.

```python
query = '''
SELECT b.title as book, 
    COUNT(DISTINCT re.review_id) AS review_count, 
    AVG(r.rating) AS average_rating
FROM books b
LEFT JOIN (
    SELECT book_id, username, review_id
    FROM reviews
) re ON b.book_id = re.book_id
LEFT JOIN (
    SELECT book_id, username, rating
    FROM ratings
) r ON b.book_id = r.book_id AND re.username = r.username
GROUP BY b.title
ORDER BY average_rating NULLS LAST
'''
user_reviews = pd.io.sql.read_sql(query, con = engine)
user_reviews
```

## Task 3
- Identify the publisher that has released the greatest number of books with more than 50 pages (this will help you exclude brochures and similar publications from your analysis).

```python
query = '''WITH filtered_books AS (
    SELECT publisher_id, 
           COUNT(book_id) as count
    FROM books
    WHERE num_pages > 50
    GROUP BY publisher_id
)
SELECT publishers.publisher, filtered_books.count
FROM filtered_books
JOIN publishers
ON filtered_books.publisher_id = publishers.publisher_id
ORDER BY count DESC
LIMIT 1'''
publisher_most_books = pd.io.sql.read_sql(query, con = engine)
publisher_most_books
```

## Task 4
- Identify the author with the highest average book rating (look only at books with at least 50 ratings).

```python
query = '''
WITH book_ratings AS (
    SELECT book_id, AVG(rating) AS avg_rating
    FROM ratings
    GROUP BY book_id
    HAVING COUNT(rating_id)>=50)
SELECT authors.author,
    AVG(book_ratings.avg_rating) AS avg_rating
FROM book_ratings
    JOIN books ON books.book_id = book_ratings.book_id
    JOIN authors ON authors.author_id = books.author_id
GROUP BY authors.author_id
ORDER BY avg_rating DESC
LIMIT 1
'''
author_highest_ratings = pd.io.sql.read_sql(
    query, con = engine)
author_highest_ratings
```

## Task 5
- Find the average number of text reviews among users who rated more than 50 books.

```python
query = '''WITH users_ratings as (
    SELECT username,
           COUNT(rating_id) as rating_count
    FROM ratings
    GROUP BY username
    HAVING COUNT(rating_id) > 50
),
users_reviews as (
    SELECT username,
           COUNT(review_id) as review_count
    FROM reviews
    WHERE username IN (
        SELECT username FROM users_ratings)
    GROUP BY username
)
SELECT AVG(review_count) as avg_reviews
FROM users_reviews
'''
average_review_count = pd.io.sql.read_sql(
    query, con = engine)
average_review_count
```
